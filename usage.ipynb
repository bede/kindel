## From Python

```python
from kindel import kindel
```

```python
consensuses, changes, report = kindel.bam_to_consensus('1.1.sub_test.bam')
```

```python
consensuses  # List of DNAIO SequenceRecords
```

```python
consensuses[0].id  # Consensus sequence ID (str)
```

```python
consensuses[0].sequence  # Consensus sequence (str)
```

```python
{c.id: str(c.sequence) for c in consensuses}  # Dict of consensus IDs and sequences
```
